```python
##### Convert RA and DEC to galactic coordinate ######
ls = []
bs = []

for i in range(len(ras)):
    c = SkyCoord(ras[i], decs[i], frame='icrs', unit='deg')
    ls.append(c.galactic.l.degree)
    bs.append(c.galactic.b.degree)
```

```python
gdms = []
gscats_s = []
for i in range(len(ls)-134):
    x = ne2001(ldeg = ls[i], bdeg=bs[i], dmd = 50, ndir=-1, verbose=False, dmd_only=False, classic=False)
    print(x[1])
    dms = x[1]['DM']
    scats_1ghz = x[1]['TAU']
    scats_600mhz = scale_scattering(x[1]['TAU'])
    
    print("This for event {} with ra {} and dec {}".format(events[i], ras[i], decs[i]))
    
    print("Galactic DM is {} pc/cc".format(dms))
    
    print("The scattering at 1GHz is {} us".format(scats_1ghz*1e3))
    
    print("The scattering at 600 MHz is {} us ".format(scats_600mhz*1e3))
    
    gdms.append(dms)
    gscats_s.append(scats_600mhz/1e3)
    
```
