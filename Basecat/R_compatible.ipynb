```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 100)
```

```python
data = pd.read_json('Basecat_morph_R.json')
data = data.fillna('')
```

```python
display(data)
```

```python
### Setting up scattering ###

fit_type = data['Fit_type'].to_numpy()

status = np.full(len(fit_type), True)

for i in range(len(fit_type)):
    if (fit_type[i] == 'Scint') or (fit_type[i] == 'Scat'):
        status[i] = False

data['Sc_status'] = status.tolist()
```

```python
#### Setting up Polarization fraction ####

pol_frac_upper = data['Pol_frac_upper'].to_numpy()
pol_frac = data['Pol_frac'].to_numpy()

status = np.full(len(pol_frac_upper), True)

comb_pol = np.full(len(pol_frac_upper), -1.0)

for i in range(len(pol_frac_upper)):
    if pol_frac_upper[i] == -1:
        status[i] = False
        
for i in range(len(pol_frac)):
    if pol_frac[i] > -1:
        comb_pol[i] = pol_frac[i]

for i in range(len(pol_frac_upper)):
    if pol_frac_upper[i] > -1:
        comb_pol[i] = pol_frac_upper[i]


comb_pol[comb_pol < 0] = 0

data['Comb_pol'] = comb_pol
data['Pol_status'] = status
```

```python
### DM, Width, Fluence, RM ###

status = np.full(len(pol_frac), False)

data['DM_status'] = status.tolist()
data['RM_status'] = status.tolist()
data['W_status'] = status.tolist()
data['Fluence_status'] = status.tolist()

data['Sc'] = data['Scattering(s)']
data['W'] = data['Min_width(s)']
data['F'] = data['Fluence(Jy_ms)']
```

```python
data.to_json('Basecat_morph_R.json')
```

```python
#### Scattering vs DM #####

data_scat = data[['Event','Sc','Sc_status','DM', 'DM_status']]
events = data_scat['Event'].to_numpy()

data_scat_true = data_scat[(data_scat['Sc_status']==True)]
print(len(data_scat_true))

print(len(data_scat))

#### Removing rows where we have scattering less than 3 times of galactic scattering ###

rm_events = np.array([25660576, 39069820, 41252817])
id_x = []
for ev in rm_events:
    i = np.where(events == ev)[0][0]
    id_x.append(i)
    
data_scat_del = data_scat.drop(id_x)
print(len(data_scat_del))

data_scat_del.to_json('Basecat_scat_dm_R.json')
```

```python
#### Scattering vs RM #####


data_scat = data[['Event','Sc','Sc_status','RM', 'RM_status', 'MW_RM']]
events = data_scat['Event'].to_numpy()

data_scat_true = data_scat[(data_scat['Sc_status']==True)]
print(len(data_scat_true))

print(len(data_scat))

#### Removing rows where we have scattering less than 3 times of galactic scattering ###

rm_events = np.array([25660576, 39069820, 41252817])
id_x = []
for ev in rm_events:
    i = np.where(events == ev)[0][0]
    id_x.append(i)
    
data_scat_del = data_scat.drop(id_x)
print(len(data_scat_del))

### Removing events with no measured RM value

data_scat_del = data_scat_del[(data_scat_del['RM']<9999)]
print(len(data_scat_del))

data_scat_del['Abs_RM_eg'] = abs(data_scat_del['RM'] - data_scat_del['MW_RM'])

data_scat_true = data_scat_del[(data_scat_del['Sc_status']==True)]
print(len(data_scat_true))

display(data_scat_del)


data_scat_del.to_json('Basecat_scat_rm_R.json')
```
