```python
#### Getting upper limit on sigma RM ########

def sigma_RM_calc(f, freqc): ### freqc in MHz
    
    lambc = 3e2/freqc #### in m
    
    print("F=", f)
    
    if (0 <= f < 1):
        sigma = np.power(lambc,-2) * np.sqrt(0.5)* np.sqrt(np.log(1/(f)))
        
    elif f >= 1:
        sigma = 0
    
    else:
        sigma = -1
    
    return sigma

sigma_RM = np.full(len(basecat_event_ids), -1, dtype=float)

for i in range(len(pol_fracs)):
    
    sig = sigma_RM_calc(pol_fracs[i], 600)
    
    print("Sigma RM is", sig, "\n")
    
    sigma_RM[i] = sig
    
```
